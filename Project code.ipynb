## Iterate the outer list and iterate through the sublist

```python
def flatten_list(_2d_list):
    flat_list = []
  
    for element in _2d_list:
        if type(element) is list:
           
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

## Reshaping of image

```python
x_train=[]
y_train=[]
for t in range(len(df)):
    folder = df["folder_name"][t]
    file1= df["files"][t]
    image = cv2.imread(os.path.join(folder, file1),cv2.IMREAD_GRAYSCALE) 
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(-1,IMG_SIZE,IMG_SIZE,1)
    image = image.reshape(image.shape[1], image.shape[2])
    if t % 500 == 0:
        print(t)
    
    x_train.append(image)
    y_train.append(df["Labels"][t])
    
```

```python
y_train
```

```python
x_train= np.array(x_train)
print("x_train has done")
y_train = np.array(pd.get_dummies(y_train), dtype= np.uint8)
```

```python
x_train = x_train.reshape(-1, 48, 48, 1)
```

```python
df=val2
```

```python
IMG_SIZE
```

```python
x_val=[]
y_val=[]
for t in range(len(df)):
    folder = df["folder_name"][t]
    file1= df["files"][t]
    image = cv2.imread(os.path.join(folder, file1),cv2.IMREAD_GRAYSCALE) 
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(-1,IMG_SIZE,IMG_SIZE,1)
    image = image.reshape(image.shape[1], image.shape[2])
    if t % 500 == 0:
        print(t)
    
    x_val.append(image)
    y_val.append(df["Labels"][t])
    
```

```python
x_val = np.array(x_val)
print("x_val has done")
y_val = np.array(pd.get_dummies(y_val), dtype= np.uint8)
```

## Tensorflow and keras import

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras.models import Sequential, Model
```

```python
from keras.layers import Dropout, Conv2D, Dense, BatchNormalization, MaxPooling2D, Flatten,Activation, Dense
```

```python
model = Sequential()

model.add(Conv2D(28, 3, 3, padding = 'same', input_shape=(x_train.shape[1:])))
model.add(Activation('relu'))
BatchNormalization(axis=1)
model.add(Conv2D(28,3,3 , padding = 'same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
BatchNormalization()
model.add(Dense(512))
model.add(Activation('relu'))
BatchNormalization()
model.add(Dense(7))
model.add(Activation('softmax'))
```
